```python
import numpy as np 
from datetime import datetime as dt 
```

```python
population_data = {'Males':0,'Females':0,'Couples':0,'Parents':0,'Baby girls':0,'Baby boys':0} 
simulation_status = 1 
```

```python
def simulation(scenario, N , MFR, FR, NGEN) : 
    q = dt.now()   
    
    ####### Calculating basic population data #######
    population_data['Males'] = int(N * MFR)
    population_data['Females'] = N - population_data['Males']
    
    print(scenario, " Scenario:\n- - - - - - - - - -")
    
    for num in range (0, NGEN) :         
        population_data['Couples'] = min(population_data['Males'], population_data['Females'])
        population_data['Parents'] = int(population_data['Couples'] * FR )
        population_data['Baby girls'] = population_data['Baby boys'] = 0

        if(scenario == 'One Child'):
            ####### Zeros in the children_arr correspond to girl, ones to boy respectively #######
            children_arr = np.random.randint(2,size = population_data['Parents'])    
            for i in children_arr.tolist() : 
                if (i == 0) : 
                    population_data['Baby girls'] += 1
                else : 
                    population_data['Baby boys'] += 1

        elif(scenario == 'One Son'):
            children_arr = np.random.randint(2, size = population_data['Parents']*5)
            counter = 0 
            
            for child in children_arr : 
                if(len(children_arr) - 4 <= counter) : 
                    break
                ## If this condition holds true, then all five children are female 
                if(1 not in children_arr[counter: counter+5]) : 
                    population_data['Baby girls'] += 5
                    counter += 5
                else : 
                    for i in range (0,5) :
                        if(children_arr[counter + i] == 0) : 
                            population_data['Baby girls'] += 1 
                        elif(children_arr[counter + i] == 1) : 
                            population_data['Baby boys'] += 1
                            counter += (5-i+1) 
                            break
                        
        else :
            raise ValueError("Unavailable scenario.")
        
        print("Generation ", num+1 ," Statistics: ")
        print("No. of MALES : ", population_data['Males'])
        print("No. of FEMALES: ", population_data['Females'])
        print("No. of couples : ", population_data['Couples'])
        print("No. of Parents : ", population_data['Parents'])
        print("No. of babies Total : ", population_data['Baby girls'] + population_data['Baby boys'])
        print("No. of baby girls : ", population_data['Baby girls'])
        print("No. of baby boys : ", population_data['Baby boys'], "\n")   
        
        ####### Entering new generation, parent data should be updated #######        
        population_data['Males'] = population_data['Baby boys']
        population_data['Females'] = population_data['Baby girls']
        
    print('--------- SURE - ',dt.now() - q, "\n")    
```

## If the user wants to end the simulation, they should enter a negative value for any of the inputs or they can provide the indicated input when they are asked whether they want to carry out another simulation.

```python
while(simulation_status) : 
    
    ########Initial Population#######
    N = int(input("Enter initial population: "))
    
    if(N <= 1) :
        raise ValueError("Initial population can't be smaller than two due to simulation purposes.")
    
    #######Male/Female ratio (to be used for new born babies’ gender)####### 
    MFR = float(input("Enter Male/Female ratio: "))
    
    if(MFR <= 0) : 
        raise ValueError("Initial Male/Female ratio can't be smaller than or equal to zero due to simulation purposes.")
    
    #######Fertility rate (the % of parents that can have a baby)#######
    FR = float(input("Enter fertility rate with as fraction value (e.g 0.4 instead of %40): "))
    
    if(FR <= 0) : 
        raise ValueError("Fertility rate can't be smaller than or equal to zero due to simulation purposes. ")
        
    #######Number of generations#######
    NGEN = int(input("Enter number of  generations: "))
    
    if(NGEN < 1) : 
        raise ValueError("Number of generations can't be smaller than 1 due to simulation purposes.")
        
    print("")        
    simulation('One Child', N, MFR, FR, NGEN)
    simulation('One Son', N, MFR, FR, NGEN)
    
    simulation_status = int(input("If you want to do another simulation enter 1, enter 0 otherwise: "))
    
    if(simulation_status != 0 and simulation_status != 1) : 
        raise ValueError("Invalid input. You should have entered either a 0 or 1.")
    
    elif(simulation_status == 0) : 
        print("\n- - - - - - - - - - Simulation completed - - - - - - - - - -")
    
```
